```python
import pandas as pd
#criando um dataframe
#dataframe = pd.DataFrame()
#criando um  dicionário de vendas durante a semana do mês 09de marmitas
vendas_marmitas ={
  'data' : ['17/09/2023', '22/09/23'],
  'lucro': [300,500],
  'empresa':['BetaLog','TagLog'],
  'quantidade':[200,300]
}
#dataframes, podem ser dicionários, tabelas e banco de dados
#criar um dataframe a partir de um dicionário
vendas_marmitas_df = pd.DataFrame(vendas_marmitas)
#display, imprime dataframe de maneira estilizada
display(vendas_marmitas_df)
```

# Nova seção

Abrindo arquivo externo

```python
vendas_df = pd.read_excel('Vendas.xlsx')
display(vendas_df)
```

```python

#mostra as 5 primeiras linhas
#se quiser colocar mais linhas, faça vendas_df.head(10) para as 10 primeiras linhas
#mostrando as 10 primeiras linhas
display(vendas_df.head(10))
#metodo shapr, mostra linhas e colunas
print(vendas_df.shape)
```

```python
#metodo describe
#metodo describe, mostra o resumo do dataframe
display(vendas_df.describe())
```

```python
#ediçao da data frame
#pegar a coluna produto
produtos = vendas_df[['Produto','ID Loja']]
#nao eh um data frame, uma serie do pandas
#serie do pandas, sao colunas do data frame
display(produtos)
```

```python
#metodo loc
#pegar 1 linha
indice=vendas_df.loc[1:5]
display(indice)
```

```python
#metodo loc
#pegar linhas que correspondem a uma condiçao

norte_shopping = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']
display(norte_shopping)
```

```python
#pegar varias linhas e colunas com loc
norte_shopping = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['ID Loja','Produto','Quantidade']]
display(norte_shopping)
```

```python
#pegar 1 linha e 1 coluna
#pegando o produto que eh camiseta, na linha 01 da coluna Produto
produto_linha_1 = vendas_df.loc[1, 'Produto']
display(produto_linha_1)
```

```python
#adicionar 1 coluna nova
#a partir de uma coluna que existe
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)
```

```python
#partir de valores padrao
vendas_df.loc[:,'Imposto'] =0
display(vendas_df)
```

```python
#adicionar 1 linha
vendas_dezembro_df = pd.read_excel('Vendas - Dez.xlsx')
display(vendas_dezembro_df)
```

```python
vendas_df = vendas_df.append(vendas_dezembro_df)
display(vendas_df)
```

```python
vendas_df_filtro = vendas_df.loc[vendas_df['Data']=='2019-12-26']
display(vendas_df_filtro)
```

```python
vendas_dez_df=vendas_df_filtro.loc[vendas_df_filtro=='2019-12-26']
display(vendas_dez_df)
```

```python
vendas_de_dezembro_df = vendas_df.loc[vendas_df['Data']=='2019-12-26',['ID Loja','Produto','Quantidade','Valor Final','Comissão']]
display(vendas_de_dezembro_df)
```

```python
comissao_dezembro=vendas_de_dezembro_df.loc[:,'Comissão'] = vendas_de_dezembro_df['Valor Final'] *0.05
display(comissao_dezembro)
```
